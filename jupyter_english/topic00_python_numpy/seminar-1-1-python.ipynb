Important hotkeys:
* Tab: autocomplete
* Shift+Tab: code hint; Shift+Tab, while holding Shift press Tab again: expanded and pinned code hint
* Ctrl+S: save current state of the notebook
* Esc: exit from input mode, after this you can mode the cursor between cells
* Shift+Enter: execute current cell + select the next cell
* Alt+Enter: execute current cell + create and select an empty code cell below

You can view all of the hotkeys by pressing Esc+H.

### 1. Standard Data Types

Why do you need standard data types?
- They simplify writing code. If there is a data strucuture that you need, it is very probable that you will find it in the Standard Library.
- They can be extended for your purposes. 
- They often are more efficient than code you implement yourself (they might be written in C and added as an extension to Python)
- They are a part of the language and if you say that you know Python, it means that you know them as well. 

Let's look on the most basic data types:
 - Numbers: `(7382, 3.14, 3+4j, Decimal, Fraction)`
 - Strings: `('net', "your's", u'joy')`
 - Lists: `([1, [2, 'three'], 4])`
 - Dictionaries: `({'Alex': 2, 'Brian': 4})`
 - Tuples: `('Leo', 21.7, 'single')`
 - Sets: `(set(1,2,3), {'a', 'b', 'c'})`
 - Files: `(open('myfile', 'r'))`
 
**Dynamic Typing**  
Variables are essentially containers for links to objects. Variables do not have a type, but links do have type. Variable are created dynamically in any part of the code using the assignment statement (`=`). 

### 2. Numbers

There are many types of numbers in Python:
 - Integers (`int`): `122`, `-4`, `99999999999`, `0o177`, `0x9ff`, `0b101010`
 - Reals (`float`): `1.0`, `3.14`, `.5`, `4E21`, `4.0e21` 
 - Complex numbers (`complex`): `3 + 4j`, `3.0 + 4.0j`,  
 - Numbers with fixed precision: `decimal.Decimal('0.1')`
 - Fractional numbers: `fractions.Fraction(3, 4)`

### 5. Loops

#### `while` loop

````python
import random

iterations = 0
while random.randint(0, 100) != 42:
    iterations += 1
    
print('{} iterations before 42 is randomly chosen'.format(iterations))
````

#### `for` loop

````python
for i in range(1, 10, 3): # Integer numbers from [1, 10) with step size 3
    print(i)
````

#### List Iteration

````python
for x in ['spam', 'eggs', 'ham']:
    if x == 'ham':
        print('yummy')
    else:
        print('not today')
````

````python
# String is (in a way) a list of chars
for i, c in enumerate('a very frequently encountered string'):
    if c == 'o':
        print(i)
````

### A bit of Data Structures

````python
[1, 2, 3, 'snake'] # List
````

````python
numbers_list = [1, 4, 2, 3, 2, 0, 5, 32]
print(numbers_list)
numbers_list.sort() # sorting changes the list
print("Sorted:", numbers_list)
numbers_list.append(322) # append also changes the list
print("Appended:", numbers_list)
````

````python
set([4, 5, 6, 6, 5, 4]) # Set is a list of unique elements
````

````python
(7, 8) # Tuple - an immutable list of fixed length
````

````python
x, y, z = (4, 5, 6) # pattern matching
print(x, y, z)
````

````python
big_numbers = dict([(9, 'nine'), (10, 'a lot')]) # Dictionary
print(big_numbers)
big_numbers[10]
````

````python
odd_squares = [number ** 2 for number in numbers_list if number % 2 == 1]
print(odd_squares) # list comprehension
````

````python
', '.join([hex(number) for number in range(16, 32) if number % 2 == 0])
````

````python
octal_numbers = {i: oct(i) for i in range(10)} # dict comprehension
print(octal_numbers)
print(5, '->', octal_numbers[5])
````
